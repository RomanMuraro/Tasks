```python
import pandas as pd
from pandas import Series, DataFrame
```

```python
movies = pd.read_table('D:\\task\\movies.dat.txt', sep = '::', names = ['MovieID', 'Title', 'Tag'],
                      engine = 'python')
movies
```

```python
ratings = pd.read_table('D:\\task\\ratings.dat.txt', sep = '::', names = ['UserID', 'MovieID', 'Rating', 'Timestap'],
                       engine = 'python')
ratings
```

```python
users = pd.read_table('D:\\task\\users.dat.txt', sep = '::', names = ['UserID', 'Gender', 'Age', 'Occepation', 'ZipCode'],
                     engine = 'python')
users
```

```python
d = pd.merge(ratings, users)
d
```

```python
d = pd.merge(d, movies)
d
```

```python
average_rating = d.pivot_table('Rating', index = 'Title', columns = 'Gender', aggfunc = 'mean')
average_rating2 = round(average_rating)
average_rating2
```

```python
average_rating_Female = average_rating2.sort_values(by = 'F', ascending = False)
m=['M']
average_rating_Female.drop(m, inplace=True, axis='columns')
average_rating_Female
```

```python
average_rating_age = d.pivot_table('Rating', index = 'Title', columns = ['Age'], aggfunc = 'mean')
average_rating_age2 = round(average_rating_age)
average_rating_age2
```

```python
average_rating_old = average_rating_age2.sort_values(by = 56, ascending = False)
average_rating_old
```
