```python
import pandas as pd
calls = [ 'name', 'gender', 'birth' ]
years = range(1880, 2011)
pieces = []
for i in years:
    df  = pd.read_table('D:\\task\\babynames\\yob%d.txt'%i, sep = ',', engine = 'python', names = calls)
    df['year'] = i
    pieces.append(df)
    data = pd.concat(pieces, ignore_index = True)
cb = data.pivot_table('birth', index = 'gender', aggfunc = 'sum')
cb  
```

```python
gender1 = data.pivot_table('birth', index = 'year', columns = 'gender', aggfunc = 'sum')
gender1
a, b = gender1['M'], gender1['F']
c = years
plt.plot(c, a, label = 'Male', color = 'green', lw = 2)
plt.plot(c, b, label = 'Female', color = 'red', lw = 2)
plt.legend()
plt.show()
```
