```python
import numpy as np
import random
import matplotlib.pyplot as plt
T = 0.8
x = int(input(' Введите размер матрицы, x = '))
k = [1, -1]
C_N = range(x)
mrx = np.zeros((x, x))
for i in range(x):
    for j in range(x):
        mrx[i, j] = random.choice(k)
print(mrx)
mas = []
mas = mrx.copy()
for m in range(int(input(' Введите кол-во итераций, m = '))):
    Sum = Sum1 = 0
    for q in range(x):
        for n in range(x):
            Sum += mrx[q][n] * mrx[q][n - 1] + mrx[q - 1][n] * mrx[q][n]           
    mas1 = []
    mas1 = mrx.copy()
    mas1[random.choice(C_N)][random.choice(C_N)] *= -1
    for g in range(x):
        for h in range(x):
            Sum1 += mas1[g][h] * mas1[g][h - 1] + mas1[g - 1][h] * mas1[g][h]
    dE = (Sum - Sum1) * -1 
    if dE <= 0:
        mrx = mas1.copy()
        mas = mrx.copy()
    else:        
        P = random.uniform(0, 1)
        W = np.exp(-dE/T)
        if P <= W:
            mas = mrx.copy()
        else:
            mrx = mas.copy()
    plt.clf()
    plt.imshow(mrx)
    plt.draw()
    plt.show()
    plt.gcf().canvas.flush_events()
plt.ioff() 
```
